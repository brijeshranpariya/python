# ***Mathematical Operations***

### Sum of arrays

```python
import numpy as np
arr1= np.random.randint(1,10,[3,3])
arr2=np.random.randint(1,10,[3,3])
arr1 + arr2
```

### Exponential of each array elements

```python
arr2 ** 2
```

### .sqrt() to find root of each array element

```python
np.sqrt(arr2)
```

### .floor() to convert each element into lowest integer value

```python
np.floor(arr2)
```

### .ceil() to convert each element into higher integer value

```python
np.ceil(arr2)
```

### .round() to find the round figure value  

```python
np.round(arr2)
np.round(arr1)
```

### .concatenate() to join two 1D arrays

```python
a=np.array([1,2,3,4])
b=np.array([4,5,6,7])
np.concatenate((b,a))
```

### .vstack() to join 2D array vertically

```python
np.vstack((arr1,arr2))
```

### .hstack() to join 2D array horizontally

```python
np.hstack((arr2,arr1))
```

### .flatten() convert multi dimensional array into 1D

```python
arr1.flatten()
```

### .split() to split 1D array into given parts

```python
a=np.array([1,2,3,4])
np.split(a,2)
```

### .hsplit() split the 2D array horizontally

```python
arr1= np.array([[1,2,3,3],[4,5,6,6]])
np.hsplit(arr1,2)
```

### .vsplit() split the 2D array vertically

```python
arr1= np.array([[1,2,3,3],[4,5,6,6]])
np.vsplit(arr1,2)
```

### .where() returns the index of given condition

```python
a = np.array([1,2,2,3,4])
np.where(a==2)
```

```python
a = np.array([1,2,2,3,4])
np.where(a>2)
```

```python
arr1= np.array([[1,2,3,3],[4,5,2,6]])
np.where(arr1==3)
```

### .sort() to sort the array

```python
np.sort(a)
```

```python
np.sort(a)[::-1] 
```

```python
np.sort(arr1)
```

```python
np.sort(arr1)[::-1]
```
